DATE

```python
df_statewide["date"] = pd.to_datetime(df_statewide["date"])
df_statewide
```

COUNTY NAME

```python
# Convert column "county_name" to string
df_statewide["county_name"] = df_statewide["county_name"].astype(str)
df_statewide["county_name"]
```

ALL BOOLEAN COLUMNS

```python
columns_boolean = [
    "raw_HA_SEARCH_PC_boolean",
    "raw_HA_SEARCH_CONCENT_boolean",
    "raw_HA_INCIDTO_ARREST_boolean",
    "raw_HA_VEHICLE_INVENT_boolean"
]

for boolean in columns_boolean:
    df_statewide[boolean] = df_statewide[boolean].map({True: 1, False: 0})

df_statewide
```

DISTRIC

```python
df_statewide["district"].value_counts()
```

VEHICLE

```python
# Show all columns name vehicle
df_statewide.filter(like="vehicle").columns
```

```python
vehicle_columns = [
    "vehicle_make",
    "vehicle_model",
    "vehicle_color",
    "vehicle_type",
]

for vehicles in vehicle_columns:
    df_statewide[vehicles] = df_statewide[vehicles].astype(str)

df_statewide["vehicle_year"] = df_statewide["vehicle_year"].astype(float)

df_statewide
```
